## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
app_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
app_df.head(2)
```

```python
# Drop the non-beneficial ID columns, 'EIN'.
app_df= app_df.drop(columns=['EIN'])
app_df.head(3)
```

```python
# Determine the number of unique values in each column.
app_df.nunique()
```

```python
# NAME counts for binning
nCounts = app_df['NAME'].value_counts()

# Choose a cutoff value and create a list of name_replace to be replaced
name_replace = list(nCounts[nCounts<5].index)

# Replace in dataframe
for name in name_replace:
    app_df['NAME'] = app_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
app_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
counts=app_df['APPLICATION_TYPE'].value_counts()
counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
app_types_to_replace = list(counts[counts<500].index)
app_types_to_replace

# Replace in dataframe
for app in app_types_to_replace:
    app_df['APPLICATION_TYPE'] = app_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
app_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
clsCounts=app_df['CLASSIFICATION'].value_counts()
clsCounts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
clsMoreThanOne = clsCounts[clsCounts>1]
clsMoreThanOne
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cls_to_replace = list (clsCounts[clsCounts<100].index)
cls_to_replace

# Replace in dataframe
for cls in cls_to_replace:
    app_df['CLASSIFICATION'] = app_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
app_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
app_df = pd.get_dummies(app_df,dtype=float)
app_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
X = app_df.drop('IS_SUCCESSFUL', axis=1).values
y = app_df['IS_SUCCESSFUL'].values
X[:3]
y[:3]

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
num_input_features = len( X_train_scaled[0])
hidden_nodes_layer1=7
hidden_nodes_layer2=14
nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=num_input_features, activation='relu'))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
# Train the model
fit_model = nn.fit(X_train_scaled,y_train,validation_split=0.15, epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
